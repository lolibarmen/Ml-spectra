```python
import pandas as pd
import numpy as np
```

```python
chem_df = pd.read_csv("Etalons_All_Bases_new.csv", sep=";", header=None)

chem_dicts = []
for _, row in chem_df.iterrows():
    sample = row[0]
    chem_data = {row[i]: float(str(row[i+2]).replace(',', '.')) for i in range(1, len(row)-2, 3)}
    chem_dicts.append({"Sample": sample, **chem_data})

chem_df_clean = pd.DataFrame(chem_dicts)
```

```python
print(chem_df_clean)
```

```python
features_df = pd.read_csv("post_componator_spectra.csv")
```

```python
print(features_df)
```

```python
merged_df = pd.merge(features_df, chem_df_clean, on="Sample", how="left")
merged_df = merged_df.dropna()
```

```python
print(merged_df)
```

```python
merged_df.to_csv("final_dataset.csv", index=False)
```
